## e1: A csv file is given to you (wine.csv)

### find the following statistics for each column (except first column)
### min, max, mean

```python
import pandas as pd

df = pd.read_csv('wine.csv', header=None)
df = df.drop([0], axis =1)
print('Min:')
print(df.min())
print('Max:')
print(df.max())
print('Mean:')
df.mean()
```

## e2: Write a Numpy based menu driven program with following options:

### option1. Enter list data
### option2. Find Maximum in the data
### option3. sort the data
### option4. find the Median of the data

```python
import numpy as np
data = []
while(1):
    option = input('enter choice')
    if option == '1':
        size = input('enter # of elements your want to enter')
        i=0
        while i<int(size):
            data.append(int(input()))
            i=i+1
    
    elif option == '2':
        print(np.array(data).max())
        
    elif option == '3':
        print(np.sort(data))
    elif option == '4':
        print(np.median(data))
        
    else:
        break
```

## e3: Year wise GDP data of three countries given below.
### years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
### country1_gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
### country2_gdp = [30.2, 240.3, 675.9, 1262.5, 3579.6, 7089.7, 10958.3]
### country3_gdp = [10.2, 27.3, 65.9, 362.5, 359.6, 789.7, 1058.3]
### Plot the line graph which shows the year wise performance of each country.

```python
from matplotlib import pyplot as plt

year = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
country1_gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
country2_gdp = [30.2, 240.3, 675.9, 1262.5, 3579.6, 7089.7, 10958.3]
country3_gdp = [10.2, 27.3, 65.9, 362.5, 359.6, 789.7, 1058.3]

plt.plot(year, country1_gdp, color='blue', marker='*', linestyle='solid')
plt.plot(year,country2_gdp, color='red', marker='o', linestyle='solid')
plt.plot(year,country3_gdp, color='green', marker='+' ,linestyle='solid')

plt.show()
```

## e4: You have iris.csv file. Perform following tasks.

```python
import pandas as pd

df = pd.read_csv('iris.csv',sep=',')

plt.scatter(df['sepal_length'],df['sepal_width'])
plt.show()

plt.scatter(df['sepal_length'],df['petal_length'])
plt.show()

plt.scatter(df['sepal_length'],df['petal_width'])
plt.show()

plt.scatter(df['sepal_width'],df['petal_length'])
plt.show()

plt.scatter(df['sepal_width'],df['petal_width'])
plt.show()

plt.scatter(df['petal_length'],df['petal_width'])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')



ax.set_xlabel('sepal width')
ax.set_ylabel('petal length')
ax.set_zlabel('patel width')
ax.set_title('Scatter Plotting')
ax.scatter(df['sepal_length'], df['petal_length'], df['petal_width'])
plt.show()
```
